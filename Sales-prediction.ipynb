```python
from sklearn import metrics

def test_evaluate(true,predicted):
    mae = metrics.mean_absolute_error(true,predicted)
    mse = metrics.mean_squared_error(true,predicted)
    rmse= np.sqrt(metrics.mean_squared_error(true,predicted))
    r2_score = metrics.r2_score(true,predicted)
    print("MAE",mae)
    print("MSE",mse)
    print("RMSE",rmse)
    print("R2_SCORE",r2_score)
    
def train_evaluate(true,predicted):
    mae = metrics.mean_absolute_error(true,predicted)
    mse = metrics.mean_squared_error(true,predicted)
    rmse= np.sqrt(metrics.mean_squared_error(true,predicted))
    r2_score = metrics.r2_score(true,predicted)
    print("MAE",mae)
    print("MSE",mse)
    print("RMSE",rmse)
    print("R2_SCORE",r2_score)
```

```python
X = df.iloc[:,:-1]
y = df["Item_Outlet_Sales"]
X.shape
```

```python
from sklearn.neighbors import LocalOutlierFactor
lof = LocalOutlierFactor()
out_col = lof.fit_predict(X)
out_col[0:6]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
                      ("std scaler",StandardScaler())
    
                                              ])

X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm_model = lm.fit(X_train,y_train)
print(lm.intercept_)
print(lm.coef_)
```

```python
pred = lm.predict(X_test)
pred
```

```python
import warnings
warnings.filterwarnings("ignore")
sns.distplot(y_test-pred)
```

```python
plt.scatter(y_test,pred)
```

```python
linear_reg = test_evaluate(y_test,pred)
linear_reg
```

```python
train_pred = lm.predict(X_train)
train_pred
```

```python
train_evaluate(y_test,pred)
```

```python
from sklearn.model_selection import cross_val_score
lm_score = cross_val_score(lm,X,y,scoring="neg_mean_squared_error",cv=10)
lm_rmse_score = np.sqrt(-lm_score)
lm_rmse_score
```

```python
lm_rmse_score.mean()
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=100, solver='cholesky', tol=0.0001, random_state=42)
ridge_model = ridge.fit(X_train,y_train)
pred2 = ridge.predict(X_test)
pred2
```

```python
sns.distplot(y_test-pred2)
```

```python
ridge_reg = test_evaluate(y_test,pred2)
ridge_reg
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1,precompute=True,selection="random",random_state=42)
lasso_model = lasso.fit(X_train,y_train)
pred3 = lasso.predict(X_test)
pred3
```

```python
sns.distplot(y_test-pred3)
```

```python
lasso_reg = test_evaluate(y_test,pred3)
lasso_reg
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=20)
dt_model = dt.fit(X_train,y_train)
pred4 = dt.predict(X_test)
pred4
```

```python
decision_tree = test_evaluate(y_test,pred4)
decision_tree
```

```python
plt.scatter(y_test,pred4)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf_model = rf.fit(X_train,y_train)
pred5 = rf.predict(X_test)
pred5
```

```python
plt.scatter(y_test,pred5)
```
