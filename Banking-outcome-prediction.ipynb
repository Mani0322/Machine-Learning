```python
req_cols = ["job","marital","education","housing","loan","contact","month","day_of_week","poutcome"]
for cols in req_cols:
    df[cols]=df[cols].astype("category")
    df[cols]=df[cols].cat.codes
df.head()
```

```python
df.drop(columns=["loan","contact","month","campaign","emp_var_rate","cons_price_idx","euribor3m","nr_employed"],inplace=True)
df.head()
```

```python
X = df.iloc[:,:-1]
y = df["Outcome"]
X.shape
```

```python
y.shape
```

```python
from sklearn.neighbors import LocalOutlierFactor
lof = LocalOutlierFactor()
outlier = lof.fit_predict(X)
outlier[0:9]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
                      ("std scaler",StandardScaler())
                          ])
X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log_model = log.fit(X_train,y_train)
pred = log.predict(X_test)
pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
c_mat = confusion_matrix(y_test,pred)
c_mat
```

```python
accuracy_score(y_test,pred)
```

```python
pred_train = log.predict(X_train)
pred_train
```

```python
accuracy_score(y_train,pred_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=203)
knn_model = knn.fit(X_train,y_train)
pred2 = knn.predict(X_test)
pred2
```

```python
c_mat = confusion_matrix(y_test,pred2)
c_mat
```

```python
accuracy_score(y_test,pred2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt_model = dt.fit(X_train,y_train)
pred3 = dt.predict(X_test)
pred3
```

```python
c_mat = confusion_matrix(y_test,pred3)
c_mat
```

```python
accuracy_score(y_test,pred3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
fr_model = rf.fit(X_train,y_train)
pred4 = rf.predict(X_test)
pred4
```

```python
c_mat = confusion_matrix(y_test,pred4)
c_mat
```

```python
accuracy_score(y_test,pred4)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm_model = svm.fit(X_train,y_train)
pred5 = svm.predict(X_test)
pred5
```

```python
c_mat = confusion_matrix(y_test,pred5)
c_mat
```

```python
accuracy_score(y_test,pred5)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(random_state=42)
gbc_model = gbc.fit(X_train,y_train)
pred6 = gbc.predict(X_test)
pred6
```

```python
c_mat = confusion_matrix(y_test,pred6)
c_mat
```

```python
accuracy_score(y_test,pred6)
```

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier()
abc_model = abc.fit(X_train,y_train)
pred7 = abc.predict(X_test)
pred7
```

```python
c_mat = confusion_matrix(y_test,pred7)
c_mat
```

```python
accuracy_score(y_test,pred7)
```
